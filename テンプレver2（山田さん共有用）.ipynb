```python
#日レベルの目標、以下のサイトにより算出（https://freeb-fis.co.jp/diet/calorie-pfc-calculator/）

##1日の栄養素（カロリー(kcal)、タンパク質P(g)、脂質F(g)、炭水化物C(g)）の目標
daily_nutrition_goals = {
    "カロリー(kcal)": 1000,
    "タンパク質P(g)": 100,
    "脂質F(g)": 100,
    "炭水化物C(g)": 100,
    }
    
##カロリー比（朝(%):昼(%):夕(%):間食(%)）：clientの生活スタイルに応じて変えていく
m_l_s_d_percentage = [30, 40, 30, 0]


#典型的な食生活
eating_habits = None

##本日の朝食、昼食、夕食、間食のメニューと栄養素データ（[[("Menu1"), (Calorie(kcal)), (Protein(g)), (Fat(g)), (Carbohydrate(g))], ...,  [("Menu#"), (Calorie(kcal)), (Protein(g)), (Fat(g)), (Carbohydrate(g))], ..., [("Menu{END}"), (Calorie(kcal)), (Protein(g)), (Fat(g)), (Carbohydrate(g))]]）。まだ摂取していない場合はNoneと定義する。
Today_Morning = None
Today_Lunch = None
Today_Dinner = None
Today_Snack = None



#顧客のクラスを定義
class Client():
    def __init__(self, daily_nutrition_goals, m_l_s_d_percentage, eating_habits, Today_Morning, Today_Lunch, Today_Dinner, Today_Snack):
        
        self.daily_nutrition_goals = daily_nutrition_goals
        self.m_l_s_d_percentage = m_l_s_d_percentage
        self.eating_habits = eating_habits

        self.Today_Morning = Today_Morning
        self.Today_Lunch = Today_Lunch
        self.Today_Dinner = Today_Dinner
        self.Today_Snack = Today_Snack

        #朝食、昼食、間食、夕食の栄養素目標を計算
        self.morning_nutrition_goal = {
            "カロリー(kcal)": int(self.daily_nutrition_goals["カロリー(kcal)"]*(m_l_s_d_percentage[0]/100)),
            "タンパク質P(g)": int(self.daily_nutrition_goals["タンパク質P(g)"]*(m_l_s_d_percentage[0]/100)),
            "脂質F(g)": int(self.daily_nutrition_goals["脂質F(g)"]*(m_l_s_d_percentage[0]/100)),
            "炭水化物C(g)": int(self.daily_nutrition_goals["炭水化物C(g)"]*(m_l_s_d_percentage[0]/100)),
            }

        self.lunch_nutrition_goal = {
            "カロリー(kcal)": int(self.daily_nutrition_goals["カロリー(kcal)"]*(m_l_s_d_percentage[1]/100)),
            "タンパク質P(g)": int(self.daily_nutrition_goals["タンパク質P(g)"]*(m_l_s_d_percentage[1]/100)),
            "脂質F(g)": int(self.daily_nutrition_goals["脂質F(g)"]*(m_l_s_d_percentage[1]/100)),
            "炭水化物C(g)": int(self.daily_nutrition_goals["炭水化物C(g)"]*(m_l_s_d_percentage[1]/100)),
            }

        self.dinner_nutrition_goal = {
            "カロリー(kcal)": int(self.daily_nutrition_goals["カロリー(kcal)"]*(m_l_s_d_percentage[2]/100)),
            "タンパク質P(g)": int(self.daily_nutrition_goals["タンパク質P(g)"]*(m_l_s_d_percentage[2]/100)),
            "脂質F(g)": int(self.daily_nutrition_goals["脂質F(g)"]*(m_l_s_d_percentage[2]/100)),
            "炭水化物C(g)": int(self.daily_nutrition_goals["炭水化物C(g)"]*(m_l_s_d_percentage[2]/100)),
            }
        
        self.snack_nutrition_goal = {
            "カロリー(kcal)": int(self.daily_nutrition_goals["カロリー(kcal)"]*(m_l_s_d_percentage[3]/100)),
            "タンパク質P(g)": int(self.daily_nutrition_goals["タンパク質P(g)"]*(m_l_s_d_percentage[3]/100)),
            "脂質F(g)": int(self.daily_nutrition_goals["脂質F(g)"]*(m_l_s_d_percentage[3]/100)),
            "炭水化物C(g)": int(self.daily_nutrition_goals["炭水化物C(g)"]*(m_l_s_d_percentage[3]/100)),
        }

        


client = Client(daily_nutrition_goals, m_l_s_d_percentage, eating_habits, Today_Morning, Today_Lunch, Today_Dinner, Today_Snack)
```
